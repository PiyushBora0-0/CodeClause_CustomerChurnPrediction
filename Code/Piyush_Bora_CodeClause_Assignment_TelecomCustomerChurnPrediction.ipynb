### Funtion to display each column that has object data type with its uniques values

# Mapping non string value to number for machine learning model training

```python
DF1['gender'].replace({'Female':1, 'Male':0},inplace =True)
```

###Replacing 'No phone service' , 'No internet service' to 'No'

```python
DF1.replace(['No phone service','No internet service'],'No',inplace=True)
```

```python
print_columns_unique_val(DF1)
```

```python
Binary_columns = ['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity',
                  'OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies',
                  'PaperlessBilling','Churn']
```

```python
for i in DF1.columns:
    if i in Binary_columns:
        DF1[i].replace({'Yes':1,'No':0},inplace=True)
```

```python
def print_unique(df):
    for i in df.columns:
        print(f"{i} : {df[i].unique()}")
```

```python
print_unique(DF1)
```

```python
df_onehot = pd.get_dummies(data = DF1,columns = ['InternetService','Contract','PaymentMethod'])
```

```python
df_onehot.head()
```

```python
print_unique(df_onehot)
```

```python
df_onehot.dtypes
```

# Scaling ['tenure',,'MonthlyCharges','TotalCharges'] in range o to 1

```python
to_scale=['tenure','MonthlyCharges','TotalCharges']
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
df_onehot[to_scale] = scaler.fit_transform(df_onehot[to_scale])
```

```python
df_onehot.sample(10)
```

```python
X = df_onehot.drop('Churn',axis='columns')
Y = df_onehot['Churn']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
print(X_train.shape," , ",X_test.shape)
```

```python
col_count=len(X_train.columns)
```

```python
col_count
```

```python
import tensorflow as tf
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Dense(col_count,input_shape=(col_count,),activation='relu'),

    keras.layers.Dense(15,activation='relu'),

    keras.layers.Dense(1,activation='sigmoid')
])
```

```python
optimizer = keras.optimizers.Adam(learning_rate=0.0001)
model.compile(optimizer = optimizer,
              loss = 'binary_crossentropy',
              metrics = ['accuracy'])
model.fit(X_train,Y_train,epochs=100)
```

```python
model.evaluate(X_test, Y_test)
```

```python
yp = model.predict(X_test)
y_pred = []
for element in yp:
    if element > 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_pred[:10]
```

```python
Y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(Y_test,y_pred))
```

```python
import seaborn as sn
cm = tf.math.confusion_matrix(labels=Y_test,predictions=y_pred)

plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
